# IMPORT LIBRARIES 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

# LOAD DATASET

```python
df = pd.read_csv(
    "Sample - Superstore.csv",
    encoding="latin1",
    engine="python"
)

df.head()
```

## DATASET INFORMATION

```python
df.info()
```

## CHECK MISSING VALUES

```python
df.isnull().sum()
```

## CONVERT DATE COLUMN

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

## CREATE DAILY SALES DATASET

```python
sales = df.groupby("Order Date")["Sales"].sum()

sales.head()
```

## PLOT SALES TREND

```python
plt.figure(figsize=(12,6))

plt.plot(
    sales,
    color='royalblue',
    linewidth=2
)

plt.title("Sales Trend", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sales")

plt.grid(True, alpha=0.3)

plt.show()
```

## PREPARE DATA FOR MACHINE LEARNING


```python
sales_df = sales.reset_index()

sales_df["Days"] = (
    sales_df["Order Date"]
    - sales_df["Order Date"].min()
).dt.days

sales_df.head()
```

## TRAIN MODEL

```python
X = sales_df[["Days"]]
y = sales_df["Sales"]

model = LinearRegression()
model.fit(X, y)
```

## PREDICT SALES

```python
sales_df["Predicted Sales"] = model.predict(X)

sales_df.head()
```

## FORECASTING GRAPH

```python
plt.style.use('ggplot')

plt.figure(figsize=(12,6))

plt.plot(
    sales_df["Order Date"],
    sales_df["Sales"],
    color='royalblue',
    linewidth=2,
    label='Actual Sales'
)

plt.plot(
    sales_df["Order Date"],
    sales_df["Predicted Sales"],
    color='crimson',
    linewidth=2,
    linestyle='--',
    label='Predicted Sales'
)

plt.title("Sales Forecasting", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## MODEL ERROR

```python
mae = mean_absolute_error(
    sales_df["Sales"],
    sales_df["Predicted Sales"]
)

print("Mean Absolute Error:", mae)
```

## FUTURE FORECASTING

```python
future_days = np.array([[1500],[1550],[1600]])

future_sales = model.predict(future_days)

print(future_sales)
```

## CREATE TIME FEATURES

```python

df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter

df[['Order Date','Year','Month','Quarter']].head()
```

## MONTHLY SALES

```python
monthly_sales = sales_df.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## CONCLUSION

The sales forecasting model was developed using historical sales data from the Superstore dataset.

Key findings:
- No missing values were found in the dataset.
- Sales trends were visualized over time.
- Time-based features were created from order dates.
- A Linear Regression model was trained to forecast future sales.
